#level 2=task 3
Task: Geographic Analysis

```python
import folium

# Step 2: Create a base map (centered at average location)
map_center = [data['Latitude'].mean(), data['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)

# Step 3: Add markers
for _, row in data.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row.get('Restaurant Name', 'Restaurant')
    ).add_to(restaurant_map)

# Step 4: Save map
restaurant_map.save("restaurant_map.html")

print("Map saved as restaurant_map.html")
```

```python
from folium.plugins import MarkerCluster

restaurant_map = folium.Map(location=map_center, zoom_start=12)
marker_cluster = MarkerCluster().add_to(restaurant_map)

for _, row in data.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row.get('Restaurant Name', 'Restaurant')
    ).add_to(marker_cluster)

restaurant_map.save("restaurant_map_cluster.html")
```

```python
restaurant_map
```

```python
from sklearn.cluster import KMeans
# Step 2: Select location data
X = data[['Latitude', 'Longitude']]

# Step 3: Apply K-Means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)

# Step 4: Plot clusters
plt.figure()

plt.scatter(data['Longitude'], data['Latitude'], c=data['Cluster'], alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,0],
            color='black', marker='X', s=100)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Restaurant Clusters")

plt.show()
```

```python

chains = data['Restaurant Name'].value_counts()
chains = chains[chains > 1].index

# Step 2: Filter only chain restaurants
chain_df = data[data['Restaurant Name'].isin(chains)]

# Step 3: Calculate average rating & votes
result = chain_df.groupby('Restaurant Name')[['Aggregate rating', 'Votes']].mean()

print(result.sort_values(by='Votes', ascending=False).head(10))
```

```python
# Step 4: Take top 5 popular chains
top_chains = result.sort_values(by='Votes', ascending=False).head(5)

# Step 5: Plot graph
plt.figure()

plt.bar(top_chains.index, top_chains['Votes'])

plt.xlabel("Restaurant Chains")
plt.ylabel("Average Votes")
plt.title("Top Restaurant Chains by Popularity")

plt.xticks(rotation=45)
plt.show()
```

# level 3=task 3
Task: Price Range vs. Online Delivery and
Table Booking

```python
import pandas as pd

# Load dataset
df = pd.read_csv("Dataset.csv")

# Step 1: Convert Yes/No to numbers
df['Has Online delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})
df['Has Table booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})

# Step 2: Group by price range
result = df.groupby('Price range')[['Has Online delivery', 'Has Table booking']].mean()

print(result)
```

```python
result.plot(kind='bar')

plt.xlabel("Price Range")
plt.ylabel("Percentage (0 to 1)")
plt.title("Price Range vs Online Delivery & Table Booking")

plt.show()
```
